# nxwlansim — MLO Throughput Demo
Demonstrates STR vs single-link throughput using the quick_scenario API.

```python
import nxwlansim as nx
print('nxwlansim version:', nx.__version__)
```

```python
# STR — 2 simultaneous links
sim_str = nx.quick_scenario(mode='str', n_links=2, n_stas=3, duration_us=200_000)
results_str = sim_str.run()
print('STR results:')
print(results_str.summary())
```

```python
# Single-link baseline
sim_single = nx.quick_scenario(mode='str', n_links=1, n_stas=3, duration_us=200_000)
results_single = sim_single.run()
print('Single-link results:')
print(results_single.summary())
```

```python
# Plot throughput comparison
results_str.plot_throughput()
```
